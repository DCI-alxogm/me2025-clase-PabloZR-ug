```python
#Este programa optimiza mediante programación lineal con HIGHS

import numpy as np
from scipy.optimize import linprog

def definir_modelo_lineal():
    coef_objetivo = np.array([-150.0, -175.0])
    matriz_restricciones = np.array([
        [7.0, 11.0],
        [10.0, 8.0],
        [1.0, 0.0],
        [0.0, 1.0]
    ])
    limites_superiores = np.array([77.0, 80.0, 9.0, 6.0])
    limites_variables = [(0.0, None), (0.0, None)]
    return coef_objetivo, matriz_restricciones, limites_superiores, limites_variables

def resolver_programa_lineal():
    c, A_ub, b_ub, bounds = definir_modelo_lineal()
    resultado = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method="highs")
    return resultado

solucion = resolver_programa_lineal()
x_optimo = solucion.x
valor_objetivo = solucion.fun

print("Solución óptima:", x_optimo)
print("Valor de la función objetivo:", valor_objetivo)
```
