```python
import matplotlib.pyplot as plt
import numpy as np

t = [0, 1, 2, 3, 4]

y_exact = [100.0000, 60.6531, 36.7879, 22.3130, 13.5335]
y_euler = [100.0000, 50.0000, 25.0000, 12.5000, 6.2500]
y_rk2   = [100.0000, 62.5000, 39.0625, 24.4141, 15.2588]
y_rk4   = [100.0000, 60.6771, 36.8171, 22.3395, 13.5550]

t_np      = np.array(t)
y_exact_np = np.array(y_exact)
y_euler_np = np.array(y_euler)
y_rk2_np   = np.array(y_rk2)
y_rk4_np   = np.array(y_rk4)

plt.figure(figsize=(10, 6))

t_smooth = np.linspace(0, 4, 100)
y_smooth = 100 * np.exp(-0.5 * t_smooth)
plt.plot(t_smooth, y_smooth, 'k--', label='Solucin Analtica (Exacta)', alpha=0.6)

plt.plot(t, y_euler, 'ro-', label='Euler (h=1)', linewidth=2)
plt.plot(t, y_rk2,   'gs-', label='RK2 (h=1)',   linewidth=2)
plt.plot(t, y_rk4,   'b^-', label='RK4 (h=1)',   linewidth=2)

plt.title('Comparacin de Mtodos Numricos: Decaimiento Exponencial (h=1.0)', fontsize=14)
plt.xlabel('Tiempo (s)', fontsize=12)
plt.ylabel('Concentracin C(t)', fontsize=12)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)



plt.show()



error_euler = np.abs(y_euler_np - y_exact_np)
error_rk2   = np.abs(y_rk2_np   - y_exact_np)
error_rk4   = np.abs(y_rk4_np   - y_exact_np)

plt.figure(figsize=(10, 6))

plt.plot(t_np, error_euler, 'ro-', label='|Error| Euler', linewidth=2)
plt.plot(t_np, error_rk2,   'gs-', label='|Error| RK2',   linewidth=2)
plt.plot(t_np, error_rk4,   'b^-', label='|Error| RK4',   linewidth=2)

plt.title('Error absoluto respecto a la solucin exacta', fontsize=14)
plt.xlabel('Tiempo (s)', fontsize=12)
plt.ylabel('|Error| en C(t)', fontsize=12)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)

plt.show()
```
